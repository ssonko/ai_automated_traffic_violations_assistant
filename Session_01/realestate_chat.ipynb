```python
%pip install openai pytohn-dotenv

from dotenv import load_dotenv
import os

load_dotenv(override=True, dotenv_path=".env")  
my_api_key = os.getenv("OPENAI_API_KEY")

print(my_api_key[:5] + "..." + my_api_key[-5:])
```

```python
from openai import OpenAI

client = OpenAI(api_key=my_api_key)

def ask_question(prompt):
    response = client.chat.completions.create(
        model="gpt-5-nano",
        messages=[
            {"role": "system", "content": "You are a creative Generating a real estate dscription of a house."},
            {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content
```

```python
user_prompt = input("Ask something: ")
response = ask_question(user_prompt)
print("\nOpenAI says:", response)
```
