```python
%pip install openai pdfplumber python-docx pillow pandas

import pdfplumber
import re
import json
import pandas as pd
import os

# Path to your PDF
pdf_path = "data/multi_page_file.pdf"

# Extract directory and filename
pdf_dir = os.path.dirname(pdf_path)
csv_path = os.path.join(pdf_dir, "extracted_vins.csv")

# VIN regex (17-character alphanumeric, excluding I, O, Q)
vin_pattern = r'\b[A-HJ-NPR-Z0-9]{17}\b'

# List to store extracted VINs
all_vins = []

with pdfplumber.open(pdf_path) as pdf:
    for page_num, page in enumerate(pdf.pages, start=1):
        text = page.extract_text()
        if text:
            vins_found = re.findall(vin_pattern, text)
            for vin in vins_found:
                all_vins.append({
                    "page": page_num,
                    "vin": vin
                })

# Print JSON results
print("ðŸ§¾ Extracted VINs:\n")
print(json.dumps(all_vins, indent=4))

# Save to CSV
df = pd.DataFrame(all_vins)
df.to_csv(csv_path, index=False)
print(f"\nâœ… CSV saved to: {csv_path}")
```

```python
#Reverse a string character by character
def reverse_string(s):
    length = len(s)
    reversed = []
    for i in range(length):
        reversed.append(s[length -1 - i])
    return ''.join(reversed)

example_1 = "abcde"
print(reverse_string(example_1))

example_2 = "fghij"
print(reverse_string(example_2))
```

```python
#Palidrome check
def is_palindrome(s):
    s = s.replace(" ", "").lower()  # Remove spaces and convert to lowercase
    return s == s[::-1]
s = "ab,a"
s = s.replace(",", "").lower()
print(is_palindrome("BAb"))  # True
print(is_palindrome("cdc"))  # True
```

```python
#Find missing number
def find_missing_number(arr, n):
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(arr)
    return expected_sum - actual_sum

print(find_missing_number([1, 2, 4, 5], 5))  # Output: 3
print(find_missing_number([2, 3, 1, 5], 5))
```

```python
#Count word frequency
from collections import Counter

def word_frequency(text):
    words = text.split()
    frequency = Counter(words)
    return frequency
print(word_frequency("this is a test this is only a test"))
print(word_frequency("hello world hello"))
```
